```python
from sklearn.ensemble import StackingClassifier
```

```python
df_X=df.iloc[:,:-1]
df_y=df.iloc[:,-1]
```

```python
df_X=sc.fit_transform(df_X)
```

```python
df_X.shape
```

```python
df_X_train, df_X_test, df_y_train, df_y_test = train_test_split(df_X, df_y, test_size=0.33)
```

```python
estimators=[('lr',LogisticRegression()),('dt',DecisionTreeClassifier()),('bn',BernoulliNB())]
```

```python
clf = StackingClassifier(estimators=estimators, final_estimator=LogisticRegression())
```

```python
clf.fit(df_X_train,df_y_train)
```

```python
final1=clf.predict(df_X_test)
```

```python
accuracy_score(df_y_test, final1)
```

```python
precision = precision_score(df_y_test, final1, average=None)
recall = recall_score(df_y_test, final1, average=None)
print('precision: {}'.format(precision))
print('recall: {}'.format(recall))
```
